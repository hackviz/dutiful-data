```python
import json
import os
import itertools
import numpy as np

import pandas as pd
```

```python
def read_releases(basedir):
    sources = []
    filenames = [os.path.join(basedir, f) for f in os.listdir(basedir) if f.endswith('.json')]
    for i, fname in enumerate(filenames):
        with open(fname) as f:
            sources.append(json.load(f))
    return sources

sources = read_releases('C:/Users/demo/Documents/notebooks/data/')

portfolios = [s['portfolios'] for s in sources]
dates = [s['date'] for s in sources]
categories = [p.split("'")[1] for folio in portfolios for p in folio]
```

```python
cat_to_id = {token: idx for idx, token in enumerate(set(categories))}
print(len(cat_to_id))
```

```python

def make_onehots(id_mappings, categories_by_x):
    arr = np.zeros((len(cat_to_id), len(categories_by_x)))
    for i, day in enumerate(categories_by_x):
        for j in day.split(','):
            if j in id_mappings.keys():
                arr[id_mappings[j], i] += 1
    return arr
```

```python
arr = make_onehots(cat_to_id, categories_by_day)
arr.shape
```

```python

frame = pd.DataFrame(arr, index=cat_to_id.keys())
```
